```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Family history Yes', 'Family history No']
female_values = [35.00, 36.00]
male_values = [36.00, 35.00]

# Bar positions
x = np.arange(len(categories))
width = 0.35  # Width of bars

# Plot
fig, ax = plt.subplots(figsize=(8, 6))
bars1 = ax.bar(x - width/2, female_values, width, label='Female', color='#FF69B4')
bars2 = ax.bar(x + width/2, male_values, width, label='Male', color='#1E90FF')

# Add value labels
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval:.2f}%', ha='center', va='bottom', fontsize=10)

# Labels, title, legend
ax.set_ylabel('Percentage')
ax.set_title('Heart Attack Prediction by Gender and Family History')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_ylim(0, max(max(female_values), max(male_values)) + 10)
ax.legend()

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Step 1: Count total and positive cases
total_counts = df.groupby(['Country', 'Sex']).size().reset_index(name='Total')
positive_counts = df[df['Heart Attack Risk'] == 1].groupby(['Country', 'Sex']).size().reset_index(name='Positive')

# Step 2: Merge and calculate percentage
merged = pd.merge(total_counts, positive_counts, on=['Country', 'Sex'], how='left').fillna(0)
merged['Heart Attack Risk (%)'] = (merged['Positive'] / merged['Total']) * 100

# Step 3: Plot
plt.figure(figsize=(14, 6))
barplot = sns.barplot(data=merged, x='Country', y='Heart Attack Risk (%)', hue='Sex',
                      palette={'Female': '#FF69B4', 'Male': '#1E90FF'})

# Step 4: Add value labels
for container in barplot.containers:
    barplot.bar_label(container, fmt='%.1f%%', label_type='edge', padding=3, fontsize=9)

# Customize
plt.title('Heart Attack Risk Percentage by Country and Sex')
plt.xlabel('Country')
plt.ylabel('Heart Attack Risk (%)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sex')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
