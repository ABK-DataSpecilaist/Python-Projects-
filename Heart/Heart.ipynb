# 18. Compare Resting Blood Pressure as per sex Column

```python
#facetgridclass is used when we want visualize distribution of a Variable or Relation between variables #
# wee have work on two columns sex which has two variables so use in facetgrid and trestbps has multiple variation so use in kdeplot
# Create the facet grid with KDE plots
g = sns.FacetGrid(data, hue="sex", aspect=4)
g.map(sns.kdeplot, 'trestbps', fill=True)

# Add legend with custom labels (assuming 0 = Female, 1 = Male)
g.add_legend(title='Sex', labels=['Female', 'Male'])

# Show the plot
plt.show()
```

# 20. Plot Continous Variables

```python
data.columns
```

```python
# there are two types of variables category Variables( Variation of value <=10) and continous Variables (Variation of value >=10)
# Lets differentiate between these two variables 
```

```python
cate_val=[]
cont_val=[]

for column in data.columns:
   if data[column].nunique() <10:
       cate_val.append(column)
   else:
       cont_val.append(column)        
```

```python
cate_val
```

```python
cont_val
```

```python
# lets plot now 
data.hist(cont_val,figsize=(15,6))
# figsize is to reduce overlap #
```

```python
# lets plot  with clear and tidy up now 
data.hist(cont_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```

```python
# lets plot Cate_val  with clear and tidy up now 
data.hist(cate_val,figsize=(15,6))
plt.tight_layout()
plt.show()
```
